```python
import requests
from pprint import pprint
import pandas as pd
import time
import json
import random
from sqlalchemy import create_engine
```

## JSON requests for SY data

```python
SY2016_resp = requests.get("https://data.cityofchicago.org/resource/fvrx-esxp.json")
SY2016_list = json.loads(SY2016_resp.text)
SY2016 = pd.DataFrame(SY2016_list)
list(SY2016.columns)
```

```python
SY2016 = SY2016[["school_id","short_name", "school_type", "primary_category", "address", "city", "state", "zip", "phone",
                "progress_report_year", "student_growth_rating", "student_attainment_rating", "growth_reading_grades_tested_pct_es",
                 "growth_math_grades_tested_pct_es", "attainment_reading_pct_es", "attainment_math_pct_es", "culture_climate_rating", "school_survey_involved_families",
                  "school_survey_supportive_environment", "school_survey_ambitious_instruction", "school_survey_effective_leaders", "school_survey_collaborative_teachers", 
                  "school_survey_safety", "suspensions_per_100_students_avg_pct", "student_attendance_avg_pct", "mobility_rate_pct", "one_year_dropout_rate_avg_pct",
                  "freshmen_on_track_school_pct_year_1", "college_enrollment_school_pct_year_1", "college_persistence_school_pct_year_2",
                   "teacher_attendance_avg_pct","school_latitude", "school_longitude"]]

SY2016["year"]= "2015-16"
len(SY2016)
```

```python
SY2017_resp = requests.get("https://data.cityofchicago.org/resource/cp7s-7gxg.json")
SY2017_list = json.loads(SY2017_resp.text)
SY2017 = pd.DataFrame(SY2017_list)
list(SY2017.columns)
```

```python
SY2018_resp = requests.get("https://data.cityofchicago.org/resource/wkiz-8iya.json")
SY2018_list = json.loads(SY2018_resp.text)
SY2018 = pd.DataFrame(SY2018_list)
list(SY2018.columns)
```

```python
SY2019_resp = requests.get("https://data.cityofchicago.org/resource/dw27-rash.json")
SY2019_list = json.loads(SY2019_resp.text)
SY2019 = pd.DataFrame(SY2019_list)
list(SY2019.columns)
```

```python
SY2019 = SY2019[["school_id","short_name", "school_type", "primary_category", "address", "city", "state", "zip", "phone", 
                 "website", "progress_report_year", "student_growth_rating", "student_attainment_rating", "growth_reading_grades_tested",
                 "growth_math_grades_tested", "attainment_reading_pct_es", "attainment_math_pct_es", "culture_climate_rating", 
                 "school_survey_involved", "school_survey_supportive", "school_survey_ambitious", 
                 "school_survey_effective", "school_survey_collaborative", "school_survey_safety", 
                 "suspensions_per_100_students", "student_attendance_avg_pct", "mobility_rate_pct", "one_year_dropout_rate_year_1",
                 "freshmen_on_track_school_1", "college_enrollment_school_1", "college_persistence_school_1",
                 "teacher_attendance_year_1","school_latitude", "school_longitude"]]
SY2019["year"]= "2018-19"
len(SY2019)
```
